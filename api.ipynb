```python
from googletrans import Translator
import pandas as pd
from os.path import exists
from tqdm import tqdm
from os import remove
```

```python
if exists('Suicide_Detection.csv'): remove('Suicide_Detection.csv')
df = pd.read_excel('Suicide_Detection.xlsx')
df.rename(columns={'Text': 'English Text'}, inplace=True)
df.drop('Unnamed: 0', 1, inplace=True)
df.info()
df.head(10)
```

```python
translator = Translator(service_urls=['translate.google.com.bn'])
eng_txt = df['English Text'].tolist()
def translate():
    for idx, sentence in tqdm(enumerate(eng_txt), total=len(eng_txt)):
        try:
            yield translator.translate(sentence, src='en', dest='bn').text
        except:
            print(f'df[{idx}] can not be translated !!!')
            yield ''
df['Bangla Text'] = list(translate())
df = df[['English Text', 'Bangla Text', 'Class']]
df.to_csv('Suicide_Detection.csv', sep=',', encoding='utf-8', index=False)
df.info()
df.head(10)
```
